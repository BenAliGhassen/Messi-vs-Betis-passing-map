```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch
```

```python
df = pd.read_csv('./messibetis.csv')

df.head()
```

```python
df.info()
```

```python
df.x = df.x * 1.2
df.y = df.y * .8

df.endX = df.endX * 1.2
df.endY = df.endY * .8
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(13.5, 8))

fig.set_facecolor('#22312b')
ax.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', axis=True, label=False, pitch_color='#22312b')
pitch.draw(ax=ax)

ax.invert_yaxis()  

for i in range(len(df.x)):
    if df.outcome[i] == 'Successful':
        ax.plot([df.x[i], df.endX[i]], [df.y[i], df.endY[i]], color='green')
        ax.plot(df.x[i], df.y[i], 'o', color='green')
    elif df.outcome[i] == 'Unsuccessful':
        ax.plot([df.x[i],df.endX[i]],[df.y[i],df.endY[i]],color='red')
        ax.plot(df.x[i], df.y[i], 'o', color='red')
```
